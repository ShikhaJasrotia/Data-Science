```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
base_dataset=pd.read_csv("C:\\Users\\156125\\Music\\Data\\application_train.csv")
```

```python
cont=base_dataset.describe().columns
cont
```

```python
for i in cont:
    x=base_dataset[i].values
    qrt1=np.quantile(x,0.25)
    qrt3=np.quantile(x,0.75)
    iqr=qrt3-qrt1
    utv=qrt3+1.5*(iqr)
    ltv=qrt1-1.5*(iqr)
    y=[]
    for i in x:
        if i<ltv or i>utv:
            y.append(np.median(x))
        else:
            y.append(i)
    base_dataset[i]=y

    
```
