```python
def null_value_treatment(base_dataset,x):
    
    print(" shape of the data before null value treatmenet",base_dataset.shape)
    
    null_values=(base_dataset.isna().sum()/base_dataset.shape[0])*100
    
    drop_columns=null_values[null_values>x].index
    
    retained_columns=null_values[null_values<x].index
    
    base_dataset.drop(drop_columns,axis=1,inplace=True)
    
    for i in base_dataset.describe().columns:
        base_dataset[i].fillna(base_dataset[i].median(),inplace=True)
    
    for i in base_dataset.describe(include='object').columns:
         base_dataset[i].fillna(base_dataset[i].value_counts().index[0],inplace=True)
    
    print(" shape of the data after null value treatmenet",base_dataset.shape)
    
    return base_dataset
    
```
