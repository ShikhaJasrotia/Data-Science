Strata Classification--->>-----Undersampling...lowest is 184 so take 184 samples of each other values

```python
x1=data_set[data_set['pclass']==1].sample(184)
x2=data_set[data_set['pclass']==2].sample(184)
x3=data_set[data_set['pclass']==3].sample(184)
```

```python
base_data_undersamples=x1.append(x2).append(x3)
```

```python
base_data_undersamples.shape
```

Strata Classification--->>>-----oversampling...highest is 491 

```python
x1=data_set[data_set['pclass']==1].sample(491,replace=True)
x2=data_set[data_set['pclass']==2].sample(491,replace=True)
x3=data_set[data_set['pclass']==3].sample(491,replace=True)
```

```python
base_data_oversamples=x1.append(x2).append(x3)
```

```python
char=[]
for i in data_set.columns:
    if i not in cont:
        char.append(i)
```

```python
char #char columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in char:
    le=LabelEncoder()
    data_set[i]=data_set[i].astype(str)
    le.fit(data_set[i])
    x=le.transform(data_set[i])
    data_set[i]=x
```

```python
for  i in data_set.columns:
    data_set[i].fillna(data_set[i].median(),inplace=True)
```
